```python
import numpy as np
import matplotlib.pyplot as plt
```

# Fermi-Dirac distribution

\begin{equation}
f_{FD}(\epsilon;\mu,T)=\frac{1}{e^{(\epsilon - \mu)/kT} + 1}
\end{equation}

Reference: Statistical and Thermal Physics Using Python - Anders Malthe-Sorenssen, Dag Kristian Dysthe

```python
def f(x):
    return 1/(np.exp(x) +1)

x = np.linspace(-10, 10, 100)
#plt.subplot(2, 1, 1)
plt.figure(figsize=(10, 5))
plt.plot(x, f(x), color='blue')
plt.xlabel("$\\frac{\epsilon - \mu}{kT}$", fontsize=14)
plt.ylabel("$f_{FD}(\epsilon;\mu,T)$", fontsize=14)
plt.show()
```

Plot of $f_{FD}(\epsilon;\mu,T)$ as a function of $\epsilon/\mu$ for different values of $T$ corresponding to $kT=0.2\mu$, $kT=\mu$ and $kT=5\mu$.

```python
def f01(x, kT):
    return 1/(np.exp((x - 1)/kT) + 1)

kT = [5, 1, 0.2]
x = np.linspace(-20, 20, 100)
plt.figure(figsize=(10, 5))
for kTval in kT:
    f = [f01(xval, kTval) for xval in x]
    plt.plot(x, f, label='kT = '+str(kTval)+'$\mu$')
    plt.legend()
    plt.xlabel("$\\frac{\epsilon}{\mu}$", fontsize=14)
    plt.ylabel("$f_{FD}(\epsilon;\mu,T)$", fontsize=14)
plt.show()
```

# Bose-Einstein distribution

\begin{equation}
f_{FD}(\epsilon_s;\mu,T)=\frac{1}{e^{(\epsilon_s - \mu)/kT} - 1}
\end{equation}

#### Plotting of Fermi-Dirac, Bose-Einstein and Maxwell-Boltzmann distribution

```python
f_FD = lambda x:1/(np.exp(x/kT) + 1)
f_BE = lambda x:1/(np.exp(x/kT) - 1)
f_MB = lambda x:1/np.exp(x/kT) 

kT = 1
distribution = ['FD', 'BE', 'MB']
plt.figure(figsize=(10, 5))
x = np.linspace(-10, 10, 100)
plt.plot(x, f_FD(x), label=distribution[0])
x = np.linspace(0.4, 10, 100)
plt.plot(x, f_BE(x), label=distribution[1])
x = np.linspace(-1, 10, 100)
plt.plot(x, f_MB(x), label=distribution[2])
plt.xlabel("$\\frac{\epsilon - \mu}{kT}$", fontsize=14)
plt.ylabel("$f_{FD}(\epsilon;\mu,T)$", fontsize=14)
plt.legend()
plt.show()
```
